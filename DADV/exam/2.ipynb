```python
import glob
import csv
import os
import pandas as pd
```

```python
for files in glob.glob('C:\\Users\\aishwarya\\Desktop\\Data Mining\\DADV\\exam\\Day\\*.csv'):
    f = open(files, 'r')
    name = os.path.basename(f.name).replace(".csv","")
    data = pd.read_csv(f)
    lis = []
    for each in range(0, len(data)-1):
        row2 = data.iloc[each]
        row1 = data.iloc[each+1]
        gl = (row2["Close"] + row1["Close"]) / row1["Close"]
        lis.append(gl)
    lis.insert(0,0)
    data.loc[:,'gainORloss'] = lis
    data.to_csv(r'C:\\Users\\aishwarya\\Desktop\\Data Mining\\DADV\\exam\\Day\\'+name+'.csv')
    f.close()
print(lis)
```
